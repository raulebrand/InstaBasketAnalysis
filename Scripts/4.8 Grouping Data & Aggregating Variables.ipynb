```python
#7a. Calculating the median of order_frequency_days for each user per request. 
#Using 'order_frequency_days' since was renamed on 4.4 task.
median_order_frequency = ords_prods_merge.groupby('user_id')['order_frequency_days'].median()
```

```python
#7b.Creating a new column 'order_frequency_flag' with default value ''
ords_prods_merge['order_frequency_flag'] = ''
```

```python
#7c. Applying the criteria and label for customers per request using new isin method due a research.
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(median_order_frequency[median_order_frequency > 20].index), 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(median_order_frequency[(median_order_frequency > 10) & (median_order_frequency <= 20)].index), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(median_order_frequency[median_order_frequency <= 10].index), 'order_frequency_flag'] = 'Frequent customer'
```

I used the "isin" method per my research to organize customers based on how often they order. Quick and smart tool that helped me check if a customer is in the group of those who order a lot, a medium amount, or not so often. This made things simpler and easier to understand in my code. "isin" made it possible to label customers without using complex steps, and it was like a friendly helper for handling a bunch of data.

```python
# Step 4: Display the counts for each order frequency category
order_frequency_counts = ords_prods_merge['order_frequency_flag'].value_counts()
order_frequency_counts
```
