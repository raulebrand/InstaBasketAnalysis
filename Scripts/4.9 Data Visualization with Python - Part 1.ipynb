```python
#Importing libraries as pd

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#creating path
path = r'C:\Users\rbrand\12-28-2024 Instacart Basket Analysis'
```

```python
#3.Importing customers.csv data set from prepared data.
df_cust = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'customers.csv'), index_col = False)
```

```python
df_cust.head()
```

```python
# Define the file path for ords_prods_merge.pkl
file_path = os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge.pkl')
```

```python
# Load the ords_prods_merge dataframe from the pickle file
ords_prods_merge = pd.read_pickle(file_path)
```

```python
# Reducing ords_prods_merge dataframe to 1,000,000 rows due memory erros
ords_prods_merge = ords_prods_merge.head(1000000)
```

```python
print("ords_prods_merge rows:", len(ords_prods_merge))
```

```python
#Using 'user_id' is the key column for merging
combined_df = pd.merge(ords_prods_merge, df_cust, on='user_id', how='left')
```

```python
combined_df.head()
```

```python
# Checking for missing values
missing_values = combined_df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Checking for duplicate rows
duplicates = combined_df.duplicated().sum()
print("Number of Duplicate Rows:", duplicates)
```

```python
# Droping rows with missing values in 'order_frequency_days' for data accurancy
combined_df = combined_df.dropna(subset=['order_frequency_days'])
```

```python
#Double checking if missing values are resolved
missing_values_after_drop = combined_df.isnull().sum()
print("Missing Values After Drop:\n", missing_values_after_drop)
```

```python
combined_df.head()
```

```python
# Droping one more column that don't add anything to analysis. I did remove two more, fam status and n_dependants bfore.
columns_to_drop = ['Unnamed: 0', 'Unnamed: 0_x']
combined_df.drop(columns=columns_to_drop, inplace=True)
```

```python
combined_df.head()
```

```python
#Exporting the combined_df DataFrame as a Pickle file.
combined_df.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'combined_df.pkl'))
```
